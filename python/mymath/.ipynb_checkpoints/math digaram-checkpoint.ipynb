```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10,0.01)
y = np.sin(x)/x
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-10,10,0.01)
y = 1/(1+np.exp(-x))
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-10,10,0.01)
y = np.sin(x*x)*x
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-0.11,0.11,0.0001)
y = np.sin(1/x)*x*x
plt.plot(x, y)
plt.show()
```
